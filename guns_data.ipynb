```python
import pandas as pd
import os
```

```python
file_path = "Resources/data_table.csv"
guns_death_rate = pd.read_csv(file_path)
# guns_death_rate = guns_death_rate.reset_index()
guns_death_rate.head()
```

```python
state_df = guns_death_rate.loc[:, ['STATE', 'YEAR', 'RATE', 'DEATHS']]
# state_df = state_df.groupby('YEAR')
state_df.head()
```

```python
columns = ['STATE', 'YEAR', 'RATE', 'DEATHS']
tx_df = state_df.loc[state_df['STATE'] == 'TX', columns]
ca_df = state_df.loc[state_df['STATE'] == 'CA', columns]
tx_vs_ca_df = pd.merge(tx_df, ca_df, how = "left", on = ["YEAR"])
tx_vs_ca_df = tx_vs_ca_df.rename(columns = {"STATE_x": "STATE", "STATE_y": "STATE", "RATE_x": "RATE", "RATE_y": "RATE", "DEATHS_x": "DEATHS", "DEATHS_y": "DEATHS"})
tx_vs_ca_df = tx_vs_ca_df.rename(index = {"STATE": "STATE"})
# xx = tx_vs_ca_df.groupby(['STATE', 'STATE'])
tx_vs_ca_df
# xx
```
