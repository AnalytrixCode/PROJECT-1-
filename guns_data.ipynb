```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import numpy as np
```

```python
file_path = "Resources/data_table.csv"
guns_death_rate = pd.read_csv(file_path)
guns_death_rate.head()
```

```python
columns = ['YEAR', 'STATE', 'RATE', 'DEATHS']
tx_vs_ca_df = guns_death_rate.loc[(guns_death_rate['STATE'] == 'TX')|(guns_death_rate['STATE'] == 'CA'), columns]
tx_vs_ca_df.head()
```

```python
tx_vs_ca = tx_vs_ca_df.loc[:, ['STATE', 'RATE', 'DEATHS']]
tx_vs_ca = tx_vs_ca.groupby('STATE').mean()
tx_vs_ca.head()
```

```python
tx_vs_ca.plot.bar()
plt.title('Texas vs California Guns-Related Death Rate and Numbers')
plt.xlabel('State')
plt.ylabel('Average Number of Death and Rate')
plt.show()
```
